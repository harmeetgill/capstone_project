To arrange the df so that each column contained a clinical_id value and each row displayed a sample - first a list of sample_ids were saved to the 'sample_id_list' variable.

Next, the .loc method and a sample_id was used to extract the corresponding row. A dictionary was constructed by zipping the clinical_id and it's corresponding value. Finally, the items in the dictionary were added to the df and the original columns were removed.

The df was saved as a csv and loaded back in. The index was reset so that the sample_id formed its own column.

```python
sample_id_list= list(grouped_clin.index)

from tqdm import tqdm
for patient in tqdm(sample_id_list):
    row = grouped_clin.loc[patient]
    dictionary={i[0]:i[1] for i in zip(row.clinical_id, row.value)}

    for x,j in dictionary.items():
        grouped_clin.loc[patient, x] = j
```

## binning variables with 'unknown' values

```python
def create_bins_with_unknowns(df, col, bins):
    df[col].replace('unknown', np.nan, inplace=True)
    df[col].replace('Unknown', np.nan, inplace=True)
    df[col]=df[col].astype(float)
    df[col]=pd.cut(df[col], bins=bins, include_lowest=True)
    df[col]=df[col].cat.add_categories('unknown')
    df[col]=df[col].fillna('unknown')
    return df[col].value_counts()
```

```python
create_bins_with_unknowns(final_df, 'TUMOR_SIZE', bins = [-0.6, 10, 20, 30, 40, 50, 60, 650])
```

```python
def create_bins(df, col, bins, na_labels):
    df[col]=df[col].astype(float)
    df[col]=pd.cut(df[col], bins=bins, include_lowest=True)
    df[col]=df[col].astype(str)
    df[col].replace('nan', na_labels,inplace=True)
    return df[col].value_counts()
```
